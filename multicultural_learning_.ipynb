#Multicultural Learning Effects on Success

##Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##Data Editing

```python
data = pd.read_csv("pisa_data.csv")
data.head()
```

```python
columns = ['SC165Q01HA', 'SC165Q02HA', 'SC165Q03HA', 'SC165Q04HA',
       'SC165Q05HA', 'SC165Q06HA', 'SC165Q07HA', 'SC165Q08HA', 'SC165Q09HA',
       'SC165Q10HA','Score','Country']
df = data[columns]
```

```python
df
```

```python
multilearning_columns = ['SC165Q01HA', 'SC165Q02HA', 'SC165Q03HA', 'SC165Q04HA',
       'SC165Q05HA', 'SC165Q06HA', 'SC165Q07HA', 'SC165Q08HA', 'SC165Q09HA',
       'SC165Q10HA']
for i in multilearning_columns:
  df[i][df[i] == 2.0] = 0.0
```

```python
df
#1.0 yes
#0.0 no
```

```python
df.isna().sum()
```

```python
#filling in the missing values
df.fillna(df.mean(),inplace=True)
```

```python
df.isna().sum()
#No more NA values. 
#Instead of Na values, we wrote the averages
```

```python
df['Sum'] = df[df.columns[0:10]].sum(axis = 1, skipna = True)
#We summed up all the values ​​and opened a new column and called it sum.
```

```python
df
```

```python
equality_0 =df['Score'][df['Sum'] == 0].mean()

more_than_1 =  df['Score'][df['Sum'] >=1].mean()
#We compared those with 0 and others as.
```

```python
equality_0
```

```python
more_than_1
```

```python
x = ['equality 0','More than 1']
y = []
y.append(equality_0)
y.append(more_than_1)
```

```python
plt.figure(figsize = (12,8))
sns.barplot(x= x , y = y,palette = 'Purples')
plt.ylim([440,470])
plt.xlabel('Total Multicultural Learning Points ')
plt.ylabel('Mean Score')
plt.title('Effect of Multicultural on Success')
plt.show()
```

##Best-Worst

```python
china = df[df['Country'] == 'B-S-J-Z (China)']
china
#Most succesfull Country: China
```

```python
china_multicult_mean = china['Sum'].mean()
china_multicult_mean
```

```python
dominican_republic=df[df["Country"]=="Dominican Republic"]
#most unsuccessful country: Dominic
```

```python
dominican_republic
```

```python
dominican_multicult_mean = dominican_republic["Sum"].mean()
dominican_multicult_mean
```

##Turkey

```python
turkey=df[df["Country"]=="Turkey"]
turkey_mean = turkey["Sum"].mean()
turkey_mean
```

##Data Visualization

```python
sns.barplot(x=[china_multicult_mean,dominican_multicult_mean,turkey_mean], y= ['China','Dominican Republic',"Turkey"], data= df,palette = 'Set3')
plt.title('reflect teachers’ practices for multicultural learning')
plt.show()
```
