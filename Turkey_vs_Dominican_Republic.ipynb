#Turkey vs Dominican Republic

##Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##Data Editing

```python
#We called our dataset and visualized it with the data.head() command.
data = pd.read_csv("pisa_data.csv")
data.head()
```

##Success Ranking

```python
result=data.groupby('Country')['Score'].mean()
```

```python
result
```

```python
worst_country = result.nsmallest(1)
worst_country
#we found the least successful country and so we can start comparing it with Turkey.
```

```python
result.plot(kind ="barh",figsize=(8,20))
plt.yticks(size = 11)
plt.xlabel("Mean Score",size = 10)
plt.ylabel("Country",size = 10)
plt.title("Most Successful Country Ranking",size = 15)
plt.show()
#by visualizing it, we saw the country ranking. We did some kind of verification.
```

```python
turkey = data[data['Country'] == 'Turkey']
```

```python
turkey.head()
```

```python
dominican_republic=data[data["Country"]=="Dominican Republic"]
```

```python
dominican_republic
```

##Turkey and Dominican Republic Comparison

#### 1-Students from socioeconomically disadvantaged homes  


```python
turkey_socioeco_mean = turkey['SC048Q03NA'].mean()
turkey_socioeco_mean
```

```python
dominican_socioeco_mean = dominican_republic["SC048Q03NA"].mean()
dominican_socioeco_mean
```

```python
sns.barplot(x=[turkey_socioeco_mean,dominican_socioeco_mean], y= ['Turkey','Dominican Republic'], data= data,palette = 'Set3')
plt.title('socioeconomically disadvantaged homes mean')
plt.show()
```

###2-Discussed their childâ€™s progress with a teacher on their own initiative

```python
care_of_parent_turkey = turkey['SC064Q01TA'].mean()
care_of_parent_turkey
```

```python
care_of_parent_dominic = dominican_republic['SC064Q01TA'].mean()
care_of_parent_dominic
```

```python
sns.barplot(x=[care_of_parent_dominic,care_of_parent_turkey], y= ['Dominican Republic','Turkey'], data= data,palette = 'Set3')
plt.title('Discussed their childâ€™s progress with a teacher on their own initiative')
plt.show()
```

###3-Government (includes departments, local, regional, state and national)

```python
goverment_turkey = turkey['SC064Q01TA'].mean()
goverment_turkey
```

```python
goverment_dominic = dominican_republic['SC064Q01TA'].mean()
goverment_dominic
```

```python
sns.barplot(x=[goverment_turkey,goverment_dominic], y= ['Turkey','Dominican Republic'], data= data,palette = 'Set3')
plt.title("State-funded school funding averages")
plt.show()
```

###4-Class Size

```python
class_size_turkey = turkey['CLSIZE'].mean()
class_size_turkey
```

```python
class_size_dominic= dominican_republic['CLSIZE'].mean()
class_size_dominic
```

```python
sns.barplot(x=[class_size_turkey,class_size_dominic], y= ['Turkey','Dominican Republic'], data= data,palette = 'Set3')
plt.title("Class Size")
plt.show()
```

###5-Approximately, how many data projectors are available in the school altogether?

```python
projector_turkey = turkey['SC004Q06NA'].mean()
projector_turkey
```

```python
projector_dominic = dominican_republic['SC004Q06NA'].mean()
projector_dominic
```

```python
sns.barplot(x=[projector_turkey,projector_dominic], y= ['Turkey','Dominican Republic'], data= data,palette = 'Set3')
plt.title("Available Projectors in school")
plt.show()
```

###6-Student-Teacher ratio

```python
ratio_turkey = turkey['STRATIO'].mean()
ratio_turkey
```

```python
ratio_dominic = dominican_republic['STRATIO'].mean()
ratio_dominic
```

```python
sns.barplot(x=[ratio_turkey,ratio_dominic], y= ['Turkey','Dominican Republic'], data= data,palette = 'Set3')
plt.title("Student-Teacher Ratio")
plt.show()
```

##CONCLUSION

1.In the first visualization, we examined the socio-economic situation (students from socioeconomically disadvantaged homes ) and commented that the socio-economic situation of the Dominican Republic was worse than that of Turkey. Perhaps one of the reasons for falling behind educationally is that they are socioeconomic disadvantaged

2.Our second visualization was about families taking care of their children.It was about whether he met the teachers of his own free will, and the Dominican Republic fell behind compared to Turkey. In addition to the reasons for being less successful than Turkey, we can also inferences that their families are less interested.

3.Where the Dominican Republic has fallen behind is the financial assistance of the government. We can say from the visuals that Turkey is more supported by the government, and we can make an inference; The reason for the difference in the success rate between them may be the government support in addition to the other ones we have mentioned.

4.When we look at the class size, we see an inverse proportion. Perhaps the size of the class(more or less students) may not have a very large contribution to the success of students.

5.Another feature that we compare is the number of school projections, and here we observed that Turkey is ahead. We can say that using technology has an impact on student success.

6.Last feature we observed is the student teacher ratio. Here we also observed that Turkey is more successful than the Dominican. The ratio of the number of student teachers can also be one of the factors affecting success. The more teachers per student, the better the success can be.
