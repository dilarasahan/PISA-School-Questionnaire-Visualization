Comparing Turkey and China which has the best score.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/pisa_data.csv')
```

```python
data.head()
```

```python
data[data['Country'] == 'Turkey']
```

First I chose some features to compare.

```python
sub = data.groupby(['Country'])['CLSIZE','CREACTIV','EDUSHORT','STAFFSHORT','Score','PROAT5AB','PROAT5AM','TOTAT'].mean().sort_values('Score')
```

```python
sub = sub.reset_index()
```

I plotted the scores by ascendig order.

```python
plt.figure(figsize = (15,15))
plt.xlim(300,590)
sns.barplot(x='Score', y='Country', data= sub,palette = 'Set2')
plt.show()
```

```python
sub_turkey = data[data['Country'] == 'Turkey']
```

```python

sub_china = data[data['Country'] == 'B-S-J-Z (China)']
```

```python
sub_china.head()
```

```python
sub_turkey.head()
```

TURKEY VS CHINA


```python
# 1-Türkiyedeki vs çindeki government funding  SC016Q01TA   
# 2-Students are grouped by ability into different classes SC042Q01TA  
# 3-Computer sayısı öğrenciler için SC004Q02TA
# 4-teaching staff in your school has attended a programme of professional development    SC025Q01NA
```

1-Turkey vs China government funding  (SC016Q01TA)

```python
turkey_funding_mean = sub_turkey['SC016Q01TA'].mean()
```

```python
china_funding_mean = sub_china['SC016Q01TA'].mean()
```

```python
x = [turkey_funding_mean,china_funding_mean]
y = ['Turkey','China']
```

```python
sns.barplot(x=x, y= y, data= data,palette = 'Set2')
plt.xlim(40,90)
plt.title('Government Funding Mean')
plt.show()
```

2- Percentage of the Schools that students are grouped by ability into different classes (SC042Q01TA)

```python
sub_china['SC042Q01TA']   # 1 = for all subjects 2 = for some  3 = not for any
```

```python
sub_china['SC042Q01TA'].value_counts()
```

```python
scch1 = list(sub_china['SC042Q01TA'].value_counts().values)[0] 
scch2 = list(sub_china['SC042Q01TA'].value_counts().values)[2]
```

```python
percentage_of_china = ((scch1+scch2) / sub_china.shape[0]) * 100
```

```python
sub_turkey['SC042Q01TA'].value_counts()
```

```python
sctr1 = list(sub_turkey['SC042Q01TA'].value_counts().values)[1] 
sctr2 = list(sub_turkey['SC042Q01TA'].value_counts().values)[2]
```

```python
percentage_of_turkey = ((sctr1 + sctr2) /  sub_turkey.shape[0]) * 100
```

```python
percentage_of_china
```

```python
percentage_of_turkey
```

```python
x1 = [percentage_of_turkey,percentage_of_china]
```

```python
sns.barplot(x=x1, y= y, data= data,palette = 'Set2')
plt.xlim(50,70)
plt.title('Students Grouped by Ability Into Different Classes Percantage')
plt.show()
```

3- Mean Sum of Computers for students (SC004Q02TA)

```python
mean_computer_china = sub_china['SC004Q02TA'].mean()
```

```python
mean_computer_turkey = sub_turkey['SC004Q02TA'].mean()
```

```python
mean_computer_china
```

```python
mean_computer_turkey
```

```python
x2 = [mean_computer_turkey,mean_computer_china]
```

```python
sns.barplot(x=x2, y= y, data= data,palette = 'Set2')
plt.xlim(0,350)
plt.title('Mean Computer Count That Available For The Students')
plt.show()
```

4- Mean teaching staff in your school has attended a programme of professional development  (SC025Q01NA)

```python
turkey_teacher_programme = sub_turkey['SC025Q01NA'].mean()
```

```python
china_teacher_programme = sub_china['SC025Q01NA'].mean()
```

```python
x3 = [turkey_teacher_programme,china_teacher_programme]
```

```python
##percentage of teaching staff in your school has attended a programme of professional development
sns.barplot(x=x3, y= y, data= data,palette = 'Set2')
plt.xlim(0,100)
plt.title('Mean Teaching Staff That Has Attended a Programme of Professional Development')
plt.show()
```

4 comparisons in 1 plot

```python
plt.subplots(nrows = 2, ncols = 2, figsize = (17,10))
plt.subplot(221)
sns.barplot(x=x, y= y, data= data,palette = 'Set2')
plt.xlim(40,90)
plt.title('Government Funding Mean')


plt.subplot(222)
sns.barplot(x=x1, y= y, data= data,palette = 'Set2')
plt.xlim(50,70)
plt.title('Students Grouped by Ability Into Different Classes Percantage')

plt.subplot(223)
sns.barplot(x=x2, y= y, data= data,palette = 'Set2')
plt.xlim(0,350)
plt.title('Mean Computer Count That Available For The Students')


plt.subplot(224)
sns.barplot(x=x3, y= y, data= data,palette = 'Set2')
plt.xlim(0,100)
plt.title('Mean Teaching Staff That Has Attended a Programme of Professional Development')
plt.suptitle('TURKEY vs. CHINA')
plt.show()

```
